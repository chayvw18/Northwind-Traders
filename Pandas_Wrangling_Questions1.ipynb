Column Operations

In the order_details table, create a new column called lineTotal = unitPrice * quantity. Show the first 10 rows with this new column.


```python
order_details['lineTotal'] = order_details['unitPrice'] * order_details['quantity']
order_details.head()
```

GroupBy Basics

Count how many orders each employee has processed (from the orders table). Display employeeID and the corresponding order count

```python
orders.groupby('employeeID').agg(empOrderCount = ('orderID', 'count')).sort_values(by='empOrderCount', ascending=False)[:3]
```

Merge orders with customers to display the orderID, orderDate, companyName, and country for each order.

```python
orders.merge(customers, how='left',
             left_on='customerID',
             right_on='customerID').loc[:, ['orderID', 'orderDate', 'companyName', 'country']]
```

```python
order_details['total_rev'] = order_details['unitPrice'] * order_details['quantity']
order_details.groupby('orderID', as_index=False)['total_rev'].sum()
```

Multi-Table Merge

Create a dataframe showing each orders orderID, orderDate, shipper companyName, and customer companyName. (Hint: join orders  shippers and orders  customers).

```python
order_full = (
    orders
    .merge(shippers, on='shipperID', suffixes=('', '_shipper'))
    .merge(customers, on='customerID', suffixes=('', '_customer'))
)
order_full.loc[:, ['orderID', 'orderDate', 'companyName', 'companyName_customer']]
```

GroupBy + Transformation

For each product (products table), calculate the average order quantity from order_details. Then merge the result back with the products table so each product row has its averageQuantityOrdered.

```python
products.assign(averageQuantityOrdered =
                products.merge(order_details, how='left', on='productID')
                .groupby('productName')['quantity'].transform('mean')
)
```
